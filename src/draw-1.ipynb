```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
name = "model-1"

l1 = np.load(name+"-train.npy")
l2 = np.load(name+"-val.npy")
l3 = np.load(name+"-test.npy")

print(np.min(l2),l3[np.argmin(l2)])
```

```python
plt.figure(figsize=(18,12))
plt.title(name+' Loss')
plt.plot(np.log(l1)[:800], color='aqua', label='Train Loss')
plt.plot(np.log(l2)[:800], color='violet', label='Valid Loss')
plt.plot(np.log(l3)[:800], color='greenyellow', label='Test Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,12))
plt.title(name+' PPL')
plt.plot(l1[:800], color='aqua', label='Train PPL')
plt.plot(l2[:800], color='violet', label='Valid PPL')
plt.plot(l3[:800], color='greenyellow', label='Test PPL')
plt.legend()
plt.show()
```
