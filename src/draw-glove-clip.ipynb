```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
name = "model-glove-clip"

l1 = np.load(name+"-train.npy")
l2 = np.load(name+"-val.npy")
l3 = np.load(name+"-test.npy")

lim = 400

print(np.min(l2[:lim]),l3[np.argmin(l2[:lim])])
```

```python
plt.figure(figsize=(18,12))
plt.title(name+' Loss')
plt.plot(np.log(l1)[:lim], color='aqua', label='Train Loss')
plt.plot(np.log(l2)[:lim], color='violet', label='Valid Loss')
plt.plot(np.log(l3)[:lim], color='greenyellow', label='Test Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(18,12))
plt.title(name+' PPL')
plt.plot(l1[:lim], color='aqua', label='Train PPL')
plt.plot(l2[:lim], color='violet', label='Valid PPL')
plt.plot(l3[:lim], color='greenyellow', label='Test PPL')
plt.legend()
plt.show()
```
